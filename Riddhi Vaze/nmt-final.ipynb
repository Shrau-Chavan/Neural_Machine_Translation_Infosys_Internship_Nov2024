# Basic Implementation

```python
from transformers import pipeline

# Load translation pipelines for different languages
translation_en_to_fr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="fr")
translation_en_to_gu = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="gu")
translation_en_to_hi = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="hi")
translation_en_to_mr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="mr")

# Function to get translation based on user choice
def translate_text():
    # Ask the user which language they want to translate to
    print("Select the target language for translation:")
    print("1. French")
    print("2. Gujarati")
    print("3. Hindi")
    print("4. Marathi")
    
    choice = input("Enter the number corresponding to your choice (1/2/3/4): ")
    
    # Ask for the input text in English
    english_text = input("Enter the text in English: ")

    # Perform translation based on the user's choice
    if choice == "1":
        translated_text = translation_en_to_fr(english_text)
        print(f"Translated Text (French): {translated_text[0]['translation_text']}")
    elif choice == "2":
        translated_text = translation_en_to_gu(english_text)
        print(f"Translated Text (Gujarati): {translated_text[0]['translation_text']}")
    elif choice == "3":
        translated_text = translation_en_to_hi(english_text)
        print(f"Translated Text (Hindi): {translated_text[0]['translation_text']}")
    elif choice == "4":
        translated_text = translation_en_to_mr(english_text)
        print(f"Translated Text (Marathi): {translated_text[0]['translation_text']}")
    else:
        print("Invalid choice. Please select a valid language option.")

# Call the function to start the translation process
translate_text()
```

# Complete Implementation

```python
import gradio as gr
from transformers import pipeline

# Load translation pipelines for different languages
translation_en_to_fr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="fr")
translation_en_to_gu = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="gu")
translation_en_to_hi = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="hi")
translation_en_to_mr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="mr")

# Function to handle translation based on user input
def translate_text(input_type, english_text, uploaded_file, target_language):
    # Determine the source text based on input type
    if input_type == "Upload a File":
        if uploaded_file is not None:
            with open(uploaded_file.name, "r", encoding="utf-8") as file:
                english_text = file.read()
        else:
            return "Please upload a valid .txt file."
    elif input_type == "Enter Text Manually":
        if not english_text.strip():
            return "Please enter text to translate."

    # Perform translation
    if target_language == "French":
        translated_text = translation_en_to_fr(english_text)
    elif target_language == "Gujarati":
        translated_text = translation_en_to_gu(english_text)
    elif target_language == "Hindi":
        translated_text = translation_en_to_hi(english_text)
    elif target_language == "Marathi":
        translated_text = translation_en_to_mr(english_text)
    else:
        return "Invalid language choice."

    return translated_text[0]['translation_text']

# Function to update visibility of input components
def update_input_visibility(input_type):
    if input_type == "Enter Text Manually":
        return gr.update(visible=True), gr.update(visible=False)
    elif input_type == "Upload a File":
        return gr.update(visible=False), gr.update(visible=True)

# Create the Gradio interface
with gr.Blocks(title="NMT Project") as iface:
    gr.Markdown("# Neural Machine Translator")
    gr.Markdown("Translate English text to French, Gujarati, Hindi, or Marathi using Neural Machine Translation models.")
    
    with gr.Row():
        input_type = gr.Radio(
            label="Choose Input Method", 
            choices=["Enter Text Manually", "Upload a File"], 
            value="Enter Text Manually"
        )
        target_language = gr.Dropdown(
            label="Select Target Language", 
            choices=["French", "Gujarati", "Hindi", "Marathi"], 
            value="French"
        )
    
    english_text = gr.Textbox(
        label="Enter Text in English", 
        placeholder="Type your text here...", 
        lines=4, 
        visible=True
    )
    uploaded_file = gr.File(
        label="Upload a .txt File", 
        file_types=[".txt"], 
        visible=False
    )
    
    translate_button = gr.Button("Translate")
    output_text = gr.Textbox(label="Translated Text", lines=4, interactive=False)

    input_type.change(
        fn=update_input_visibility, 
        inputs=[input_type], 
        outputs=[english_text, uploaded_file]
    )
    translate_button.click(
        fn=translate_text, 
        inputs=[input_type, english_text, uploaded_file, target_language], 
        outputs=output_text
    )

iface.launch()
```

# **Final Implementation**

Can translate .txt, .docx, .pdf files

```python
import gradio as gr
from transformers import pipeline
from docx import Document  # To handle .docx files
import PyPDF2  # To handle .pdf files

# Load translation pipelines for different languages
translation_en_to_fr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="fr")
translation_en_to_gu = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="gu")
translation_en_to_hi = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="hi")
translation_en_to_mr = pipeline("translation", model="facebook/m2m100_418M", src_lang="en", tgt_lang="mr")

# Function to read file content based on file type
def read_file_content(file):
    if file.name.endswith(".txt"):
        with open(file.name, "r", encoding="utf-8") as f:
            return f.read()
    elif file.name.endswith(".docx"):
        doc = Document(file.name)
        return "\n".join([para.text for para in doc.paragraphs])
    elif file.name.endswith(".pdf"):
        pdf_reader = PyPDF2.PdfReader(file.name)
        return "\n".join([page.extract_text() for page in pdf_reader.pages])
    else:
        return "Unsupported file format."

# Function to handle translation based on user input
def translate_text(input_type, english_text, uploaded_file, target_language):
    # Determine the source text based on input type
    if input_type == "Upload a File":
        if uploaded_file is not None:
            english_text = read_file_content(uploaded_file)
        else:
            return "Please upload a valid .txt, .docx, or .pdf file."
    elif input_type == "Enter Text Manually":
        if not english_text.strip():
            return "Please enter text to translate."

    # Perform translation
    if target_language == "French":
        translated_text = translation_en_to_fr(english_text)
    elif target_language == "Gujarati":
        translated_text = translation_en_to_gu(english_text)
    elif target_language == "Hindi":
        translated_text = translation_en_to_hi(english_text)
    elif target_language == "Marathi":
        translated_text = translation_en_to_mr(english_text)
    else:
        return "Invalid language choice."

    return translated_text[0]['translation_text']

# Function to update visibility of input components
def update_input_visibility(input_type):
    if input_type == "Enter Text Manually":
        return gr.update(visible=True), gr.update(visible=False)
    elif input_type == "Upload a File":
        return gr.update(visible=False), gr.update(visible=True)

# Create the Gradio interface
with gr.Blocks(title="NMT Project", css="""
    .gradio-container {
        background-color: lightyellow;
    }
    .gradio-container .gradio-row {
        background-color: lightyellow;
        border: 2px solid black;
        padding: 10px;
    }
    .gradio-container .gradio-textbox, .gradio-container .gradio-dropdown, .gradio-container .gradio-radio, .gradio-container .gradio-file, .gradio-container .gradio-button {
        background-color: lightyellow;
        border: 2px solid black;
        padding: 10px;
    }
    h1 {
        font-weight: bold;
        text-align: center;
    }
    p {
        text-align: center;
    }
""") as iface:
    # Heading and description with bold and centered text
    gr.Markdown("<h1>Neural Machine Translator</h1>")
    gr.Markdown("<p>Translate English text to French, Gujarati, Hindi, or Marathi using Neural Machine Translation models.</p>")
    
    with gr.Row():
        input_type = gr.Radio(
            label="Choose Input Method", 
            choices=["Enter Text Manually", "Upload a File"], 
            value="Enter Text Manually"
        )
        target_language = gr.Dropdown(
            label="Select Target Language", 
            choices=["French", "Gujarati", "Hindi", "Marathi"], 
            value="French"
        )
    
    english_text = gr.Textbox(
        label="Enter Text in English", 
        placeholder="Type your text here...", 
        lines=4, 
        visible=True
    )
    uploaded_file = gr.File(
        label="Upload a .txt, .docx, or .pdf File", 
        file_types=[".txt", ".docx", ".pdf"], 
        visible=False
    )
    
    translate_button = gr.Button("Translate")
    output_text = gr.Textbox(label="Translated Text", lines=4, interactive=False)

    input_type.change(
        fn=update_input_visibility, 
        inputs=[input_type], 
        outputs=[english_text, uploaded_file]
    )
    translate_button.click(
        fn=translate_text, 
        inputs=[input_type, english_text, uploaded_file, target_language], 
        outputs=output_text
    )

iface.launch()
```
